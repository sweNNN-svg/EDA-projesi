 ### Satışların Toplamı

```python
sales_g = data.Global.sum()
sales_n = data["North America"].sum()
sales_e = data.Europe.sum()
sales_j = data.Japan.sum()
sales_r = data["Rest of World"].sum()


print("Global Satışlar: {}".format(sales_g))
print("North America Satışlar: {}".format(sales_n))
print("Europe Satışlar: {}".format(sales_e))
print("Japan Satışlar: {}".format(sales_j))
print("Rest of World Satışlar: {}".format(sales_r))
```

### Görselleştirme 1: Global Sales by Region

```python
labels = data.describe().columns[1:].tolist()
labels
```

```python
values = [sales_n, sales_e, sales_j, sales_r, sales_g]
values
```

```python
plt.bar(labels, values, color="b")
plt.xlabel("Region")
plt.ylabel("Billion $")
plt.title("Global Sales by Region")

plt.show()
```

### Görselleştirme 2: Global Sales by Region

```python
labels = data.describe().columns[1:-1].tolist()
values = [sales_n, sales_e, sales_j, sales_r]
plt.pie(values, labels=labels, autopct = "%1.1f%%")
plt.title("Global Sales by Region")
plt.show()
```

### Görselleştirme 3: Global Sales by Year

```python
# Şimdi amacımız yıllara göre toplam satışı bulmak.
```

```python
data.Year.value_counts()
```

```python
sales_by_year = data.groupby("Year")["Global"].sum()
print(sales_by_year)
```

```python
sales_by_year.plot(kind = "bar")
plt.xlabel("Year")
plt.ylabel("Global Sales by Year")
plt.show()
```

### Görselleştirme 4: Region Sales by Year

```python
sales_by_year = data.groupby("Year")["North America"].sum()
sales_by_year.plot(kind = "bar", color="black")
plt.xlabel("Year")
plt.ylabel("North America Sales by Year")
plt.show()
```

```python
sales_by_year = data.groupby("Year")["Japan"].sum()
sales_by_year.plot(kind = "bar", color="red")
plt.xlabel("Year")
plt.ylabel("Japan Sales by Year")
plt.show()
```

```python
sales_by_year = data.groupby("Year")["Europe"].sum()
sales_by_year.plot(kind = "bar", color="brown")
plt.xlabel("Year")
plt.ylabel("Europe Sales by Year")
plt.show()
```

```python
sales_by_year = data.groupby("Year")["Rest of World"].sum()
sales_by_year.plot(kind = "bar", color="yellow")
plt.xlabel("Year")
plt.ylabel("Rest of World Sales by Year")
plt.show()
```

```python
sales_of_games = data.head(10).groupby("Game")["Global"].sum().sort_values(ascending=False)
sales_of_games.plot(kind = "bar", color="blue")
plt.xlabel("Game")
plt.ylabel("Global Sale (Billion $)")
plt.title("Global Sales by Year")
plt.show()
```
